##### this empty vector will be filled with a mean at each simulation sample done:

```r
srs <- function(iters,data,n,comp_mean="False"){ means=rep(NA,iters)
                                        sds = rep(NA, iters)
                                            
    for(i in 1:iters){if(comp_mean=="True"){samp1=sample(data,n)
                                     means[i]<-mean(samp1)
                                    }
                     if(comp_mean=="False"){samp2=sample(data,n)
                                    sds[i]<-sd(samp2)
                                          }}
                                                if(comp_mean=="True"){hist(means)
                                                                     abline(v=mean(data), lty=4,lwd=4,col="red")}
                                                if(comp_mean=="False"){hist(sds)
                                                                       abline(v=sd(data), lty=4,lwd=4,col="green")}
                                                
                                                output=list(sd(data),mean(sds),mean(data),mean(means))
                                                names(output)<-c("real sd","mean sd","real mean","mean of means")

                                                
return(output)}
```

```r
srs(iters=1000,data=apipop$ell,n=100,comp_mean="True")
```

```r
sampling <-  function(iters,data,n,plotit="False"){ means=rep(NA,iters)
                                        sds = rep(NA, iters)
                                                     for(i in 1:iters){s1=sample(data,n)
                                                                      s2=sample(data,n)
                                                                      
                                                                      means[i]<- mean(s1)
                                                                      sds[i]<- mean(s2)}
                                                   
                                                     if(plotit=="True"){hist(means)
                                                                        abline(v=mean(data),lwd=4,lty=4,col="red")
                                                                        
                                                                        hist(sds)
                                                                        abline(v=sd(data),lwd=4,lty=4,col="green")}
                                                   output=list(sd(data),mean(sds),mean(data),mean(means))
                                                   names(output)<-c("real sd","mean sd","real mean","mean of means")
                                                   return(output)
                                                     }
```

```r
sampling(iters=1000,data=apipop$ell,n=100,plotit="True")
```
